## Task 1. Generate some toy data to test the poisson-GPFA code

As usual, we start by verifying our code on toy data. The cell below already contains code to generate data for 30 neurons, 100 trials (1000 ms each) and 50ms bin size.
The neurons' firing rate $\lambda_k$ is assumed to be a constant $d_k$ modulated by a one-dimensional latent state $x$, which is drawn from a Gaussian process:

$\lambda_k = \exp(c_kx + d_k)$

Each neuron's weight $c_k$ is drawn randomly from a normal distribution and spike counts are sampled form a Poisson distribution with rate $\lambda_k$.

Your task is to fit a Poisson GPFA model with one latent variable to this data (see `engine.PPGPFAfit`).

*Grading: 3 pts*

````python
# Initialize random number generator
np.random.seed(123)
# Specify dataset & fitting parameters
xdim      =1#latent dimensionality to fit
ydim      = 30		 #number of neurons in the dataset
numTrials = 100		
trialDur  = 1000 # in ms
binSize   = 50	 # in ms
maxEMiter = 100		
dOffset   = 1	 # controls firing rate


# Sample from the model (make a toy dataset)
training_set  = util.dataset(
    seed      = 345533,
    xdim      = xdim,
	ydim      = ydim,
	numTrials = numTrials,
	trialDur  = trialDur,
	binSize   = binSize,
	dOffset   = dOffset,
	fixTau 	  = True, 
	fixedTau  = np.array([0.1]),
	drawSameX = False) ##np.random.randint(10000) 345533 #np.random.randint(10000)
````

### Fit the model

````python
# Initialize parameters using Poisson-PCA
initParams = util.initializeParams(xdim, ydim, training_set)

# Fit using vanilla (batch) EM
fitToy = engine.PPGPFAfit(
	experiment 		= training_set,
	initParams 		= initParams,
	inferenceMethod = 'laplace',
	EMmode 			= 'Batch',
	maxEMiter 		= 100)
````

````python
#some useful functions
def allTrialsState(fit,p):
    """Reshape the latent signal and the spike counts"""
    x = np.zeros([p,0])
    for i in range(len(fit.infRes['post_mean'])):
        x = np.concatenate((x,fit.infRes['post_mean'][i]),axis=1)
    return x

def allTrialsX(training_set):
    """Reshape the ground truth 
    latent signal and the spike counts"""
    x_gt = np.array([])
    for i in range(len(training_set.data)):
        x_gt =  np.concatenate((x_gt,training_set.data[i]['X'][0]),axis = 0)
    return x_gt


def cov_analytical(fit):
    """Analytical covariance of the model
     r = exp(cX+d)
     y ~ Poiss(r) """
    d = fit.optimParams['d']
    rho = d
    C = fit.optimParams['C']
    n = len(rho)
    lamb = np.dot(C,C.T)

    E_y = np.exp(0.5*np.diag(lamb)+rho) # E[lambda]
    
    # diag(E[lambda]) + Cov[lambda]
    return (((np.eye(n)*E_y) + np.exp(lamb)*E_y[:,np.newaxis].dot(E_y[np.newaxis,:]))
                                -E_y[:,np.newaxis].dot(E_y[np.newaxis,:]) )
````

````python
# extract latent variables, fit and ground truth
x = allTrialsState(fitToy,1)
x_gt = allTrialsX(training_set)

# calculate Cov analytically from fit
CoV = cov_analytical(fitToy)

# bounds for plotting
vmi = np.min([np.cov(training_set.all_raster),CoV])
vma = np.max([np.cov(training_set.all_raster),CoV])

gridspec.GridSpec(6,3)
size = (8,3)
plt.figure(figsize=[10,10])
plt.subplot2grid(size, (0,0), colspan=1, rowspan=2)
plt.scatter( x =-training_set.params['C'],y = fitToy.optimParams['C'])
plt.plot([-0.6,1],[-0.6,1],ls=':',color='black')
plt.axis('equal')
plt.xlabel('true')
plt.ylabel('inferred')
plt.title('Weights')

plt.subplot2grid(size, (0,2), colspan=1, rowspan=2)
plt.scatter(x = training_set.params['d'],y = fitToy.optimParams['d'])
plt.plot([-0.8,1],[-0.8,1],ls=':',color='black')

plt.xlabel('true')
plt.ylabel('inferred')
plt.title('Bias')

plt.subplot2grid(size, (2,0), colspan=1, rowspan=2)
sns.heatmap(fitToy.E_yy_obs,vmin = vmi,vmax = vma)
#sns.heatmap(np.cov(training_set.all_raster), vmin = vmi,vmax = vma)
plt.grid('off')
plt.title('Spike Counts Covariance')

plt.subplot2grid(size, (2,2), colspan=1, rowspan=2)
#sns.heatmap(fitToy.E_yy_optim_params,vmin = vmi,vmax = vma )
sns.heatmap(CoV,vmin = vmi,vmax = vma )
plt.grid('off')
plt.title('Rank-one approxiamtion')


plt.subplot2grid(size, (4,0), colspan=4, rowspan=3)
#one trial latent state
#plt.plot(fitToy.infRes['post_mean'][0][0],label = 'estimated')
#plt.plot(training_set.data[0]['X'].T,label = 'ground truth')

plt.plot(-x.T[0:400],label ='Estimated')
plt.plot(x_gt[0:400],label = 'ground truth')
plt.xlabel('Time (bins)')
plt.title('latent state')
plt.title('Latent state (20 trials)')
plt.axvline(0, ls=':',c='black',label='trial boundaries')

for i in range(1,20):
    plt.axvline(i*20, ls=':',c='black')
plt.legend()

plt.tight_layout()
````

## Task 4. Visualization of correlation matrix.
Plot (a) the total covariance matrix of the data as well as its approximation using (b) one and (c) five latent variable(s).
Use the analytical solution for the covariance matrix of the approximation*. 
Note that the solution is essentionally the mean and cov of the [log-normal distribution](https://en.wikipedia.org/wiki/Log-normal_distribution).

$ \mu = \exp(\frac{1}{2} \text{ diag}(CC^T)+d)\\
\text{Cov}= \mu\mu^T \exp(CC^T)+\text{ diag}(\mu) - \mu\mu^T$ 

*[Krumin, M., and Shoham, S. (2009). Generation of Spike Trains with Controlled Auto- and Cross-Correlation Functions. Neural Computation 21, 1642–1664](http://www.mitpressjournals.org/doi/10.1162/neco.2009.08-08-847).

*Grading 3 pts*

````python
def cov_analytical(fit):
    """Analytical covariance of the model
     r = exp(cX+d)
     y ~ Poiss(r) """
    d = fit.optimParams['d']
    C = fit.optimParams['C']
    n = len(d)
    cct = np.dot(C,C.T)

    mu = np.exp(0.5*np.diag(cct)+d)
    return (( np.exp(cct) * mu[:,np.newaxis].dot(mu[np.newaxis,:]))
            +(np.eye(n)*mu) 
            -mu[:,np.newaxis].dot(mu[np.newaxis,:]) )


# sample covariance
obs_corr = np.cov(data.all_raster)

# analytical computed on fits
opt_corrR1 = cov_analytical(xval.fits[0])
opt_corrR6= cov_analytical(xval.fits[4])

# set color range
vmin = -1 #-1 #np.min([obs_corr,opt_corrR1,opt_corrR6])
vmax = 1 #1 #np.max([obs_corr,opt_corrR1,opt_corrR6])

plt.figure(0,figsize = [15,5])
plt.subplot(1,3,1)
plt.imshow(obs_corr,interpolation="nearest",cmap ='bwr',vmin = vmin,vmax = vmax)
plt.colorbar(**{'shrink':0.7 })
plt.title('Sample covariance')

plt.subplot(1,3,2)
plt.imshow(opt_corrR1,interpolation="nearest",cmap ='bwr',vmin = vmin,vmax = vmax)
plt.title('Rank-one approximation ')
plt.colorbar(**{'shrink':0.7 })
plt.subplot(1,3,3)
plt.imshow(opt_corrR6,interpolation="nearest",cmap ='bwr',vmin = vmin,vmax = vmax)
plt.colorbar(**{'shrink':0.7 })
plt.title('Rank-five approximation ')

````
