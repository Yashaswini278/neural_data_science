## Load data

````python
spikes = pd.read_csv('../data/nda_ex_5_spiketimes.csv') # neuron id, spike time
stims = pd.read_csv('../data/nda_ex_5_stimulus.csv')  # stimulus onset in ms, direction

stimDur = 2000.    # in ms
nTrials = 11       # number of trials per condition
nDirs = 16         # number of conditions
deltaDir = 22.5    # difference between conditions

stims['StimOffset'] = stims['StimOnset'] + stimDur
````

We require some more information about the spikes for the plots and analyses we intend to make later. With a solution based on dataframes, it is natural to compute this information here and add it as additional columns to the `spikes` dataframe by combining it with the `stims` dataframe. We later need to know which condition (`Dir`) and trial (`Trial`) a spike was recorded in, the relative spike times compared to stimulus onset of the stimulus it was recorded in (`relTime`) and whether a spike was during the stimulation period (`stimPeriod`). 

````python
# init some arrays
cond = np.empty_like(spikes['SpikeTimes'])        
cond[:] = np.NAN
relTime = cond.copy()     
trial = cond.copy()

stimPeriod = np.zeros_like(spikes['SpikeTimes'],dtype=bool)       

dirs = np.unique(stims['Dir'])
trialCounter = np.zeros_like(dirs)

for i, row in stims.iterrows():
    trialCounter[row['Dir']==dirs] +=1 
    
    # we want to plot also spike a little before and after stimulation
    i1 = spikes['SpikeTimes'] > row['StimOnset']-100 
    i2 = spikes['SpikeTimes'] < row['StimOffset']+100
 
    idx = i1.get_values() & i2.get_values()
    
    # direction of the bar for each spike
    cond[idx] = row['Dir']      
    
    # spike time relative to stimulus onset
    relTime[idx] = spikes.loc[idx,'SpikeTimes'] - row['StimOnset'] 
    
    # repetition of a certain direction
    trial[idx] = trialCounter[row['Dir']==dirs]
    
    # but we also want to keep track of what is inside the stimulation period
    i1 = spikes['SpikeTimes'] > row['StimOnset']
    i2 = spikes['SpikeTimes'] < row['StimOffset']
    stimPeriod[i1.get_values() & i2.get_values()] = True      
    
    
spikes['Dir'] = cond
spikes['relTime'] = relTime
spikes['Trial'] = trial
spikes['stimPeriod'] = stimPeriod

spikes = spikes.dropna()

spikes.head()
````

````python
spikes.to_csv('spikes.csv')
````

## Task 1: Plot spike rasters

In a raster plot, each spike is shown by a small tick at the time it occurs relative to stimulus onset. Implement a function `plotRaster()` that plots the spikes of one cell as one trial per row, sorted by conditions (similar to what you saw in the lecture). Why are there no spikes in some conditions and many in others?

If you opt for a solution without a dataframe, you may need to change the interface of the function.

*Grading: 2 pts*


````python
def plotRaster(spikes, neuron):
#    Plot spike rasters for a single neuron sorted by condition
#
#    spikes    pandas dataframe with columns 
#                  Neuron
#                  SpikeTimes
#                  Dir
#                  relTime
#                  Trial
#                  stimPeriod
#    neuron    scalar

    spk = spikes[spikes['Neuron']==neuron]

    dirs = np.unique(spikes['Dir'])
    trials = np.unique(spikes['Trial'])

    plt.figure(figsize=(5,8))

    row = 0
    for x in itertools.product(dirs,trials):
        i1 = spk['Dir'] == x[0]
        i2 = spk['Trial'] == x[1]
        t = spk.loc[i1.get_values() & i2.get_values() ,'relTime']
        plt.plot(t,row * np.ones_like(t),'k.', markersize=3)
        row += 1

    plt.xlim([-100,stimDur+100])
    plt.ylim([0, row])

    # make the lines separate the different conditions
    plt.yticks(np.concatenate(([0],np.cumsum(trials[-1]*np.ones(len(dirs)-1)))),dirs)

    plt.xlabel('Time (ms)')
    plt.ylabel('Trial (sorted by direction)')
````

Show examples of different neurons. Good candidates to check are 28, 29 or 37. 

````python
plotRaster(spikes,28)
plotRaster(spikes,29)
plotRaster(spikes,37)
````

## Task 2: Plot spike density functions

Compute an estimate of the spike rate against time relative to stimulus onset. There are two ways:
* Discretize time: Decide on a bin size, count the spikes in each bin and average across trials. 
* Directly estimate the probability of spiking using a density estimator with specified kernel width. 

Implement one of them in the function `plotPsth()`.


*Grading: 2 pts*


````python
def plotPSTH(spikes, neuron):
#    Plot PSTH for a single neuron sorted by condition
#
#    spikes    pandas dataframe with columns 
#                  Neuron
#                  SpikeTimes
#                  Dir
#                  relTime
#                  Trial
#                  stimPeriod
#    neuron    scalar

    dt = 10
 
    spk = spikes[spikes['Neuron']==neuron]

    dirs = np.unique(spikes['Dir'])
    trials = np.unique(spikes['Trial'])

    plt.figure(figsize=(5,8))

    bins = np.arange(-100,2100,dt)
    bin_centers = bins[0:-1] + dt/2
    
    x = np.arange(-100,2100)

    expCount = np.max(spk['stimPeriod'].groupby([spikes['Dir']]).count()) / len(bins)

    offset = np.zeros(len(dirs))

    for i, dir in enumerate(dirs):
        i1 = spk['Dir'] == dir
        t = spk.loc[i1.get_values() ,'relTime']

        offset[i] = 3*i*expCount
        
        # binned PSTH
        h = np.histogram(t,bins=bins)
        plt.plot(bin_centers,h[0]+offset[i],'k',linewidth=1)

    plt.xlim([-100,stimDur+100])
    plt.ylim([0, np.max(offset)+3*expCount])

    # make the lines separate the different conditions
    plt.yticks(offset,dirs)

    plt.xlabel('Time (ms)')
````

## Task 3: Fit and plot tuning functions

The goal is to visualize the activity of each neuron as a function of stimulus direction. First, compute the spike counts of each neuron for each direction of motion and trial.  The result should be a matrix `x`, where $x_{jk}$ represents the spike count of the $j$-th response to the $k$-th direction of motion (i.e. each column contains the spike counts for all trials with one direction of motion).	If you used dataframes above, the `groupby()` function allows to implement this very compactely. Make sure you don't loose trials with zero spikes though.

Fit the tuning curve, i.e. the average spike count per direction, using a von Mises model. To capture the non-linearity and direction selectivity of the neurons, we will fit a modified von Mises function:

$$ f(\theta) = \exp(\alpha + \kappa (\cos (2*(\theta-\phi)))-1) + \nu (\cos (\theta-\phi)-1))$$

Here, $\theta$ is the stimulus direction. Implement the von Mises function in `vonMises()` and plot it to understand how to interpret its parameters $\phi$, $\kappa$, $\nu$, $\alpha$. Perform a non-linear least squares fit using a function of your choice. Implement the fitting in `tuningCurve()`. 

Plot the average number of spikes per direction, the spike counts from individual trials as well as your optimal fit.

Select two cells that show nice tuning to test your code.

*Grading: 3 pts*

````python
def vonMises(theta, alpha, kappa, nu, phi):
# Evaluate the parametric von Mises tuning curve with parameters p at locations theta. 
# The locations are given in degrees, not radians.
#
#   theta:  1 by N
#   alpha, kappa, nu, phi:   parameters of the function
#
#   f:      1 by N

    theta = theta / 180 * np.pi
    f = np.exp(alpha + kappa * (np.cos(theta - phi/180*np.pi) - 1) +
            nu * (np.cos(2 * (theta - phi/180*np.pi)) - 1));

    return f
    
````

Plot the von Mises function while varying the parameters systematically.

````python
x = np.arange(0,360)

p = [2, 3, 3, 120]
plt.plot(x,vonMises(x, p[0], p[1], p[2], p[3]))

p = [2, 3, 3, 90] # last parameter varies location of peak
plt.plot(x,vonMises(x, p[0], p[1], p[2], p[3]))

p = [1, 3, 3, 180] # first parameter varies height of the peak
plt.plot(x,vonMises(x, p[0], p[1], p[2], p[3]))

p = [1, 3, 0, 150] # second parameter control 1st fourier mode
plt.plot(x,vonMises(x, p[0], p[1], p[2], p[3]))

p = [1, 0, 3, 150] # third parameter control 2nd fourier mode
plt.plot(x,vonMises(x, p[0], p[1], p[2], p[3]))

plt.xlabel('Direction (deg)')
plt.ylabel('Spike count')
plt.xlim((0,360))
plt.legend(['2,3,3,120','2,3,3,90', '1,3,3,180', 
            '1,3,0,150', '1,0,3,150' ],loc=2, bbox_to_anchor=(1.05, 1))
````

````python
def tuningCurve(counts, dirs, show):
    '''
    Fit a von Mises tuning curve to the spike counts in count with direction dir using a least-squares fit.
    Plot the data if show is True, otherwise just return the fit.
        count      #trials x 1, contains the spike count during the stimulation period
        dirs       #trials x 1, contains the direction in degrees
        show       Boolean, indicates whether a plot is generated

        p          1 by 4 parameter vector of tuning curve function
    '''


    df = pd.DataFrame({'dirs': dirs, 'counts': counts})

    df_mean = df.groupby('dirs').mean().reset_index()
    x = df_mean['dirs']
    y = df_mean['counts']


    p = [2, 3, 4, x[np.argmax(y)]]

    p = opt.curve_fit(vonMises, x, y, p0=p)[0]


    if show:
        sns.swarmplot(data=df,x='dirs',y='counts',palette='Blues_d', alpha=.5)
        sns.pointplot(data=df,x='dirs',y='counts',palette='Blues_d')
        
        xx = np.linspace(0,360,100)
        plt.plot(xx/np.max(xx)*nDirs,vonMises(xx, p[0], p[1], p[2], p[3]))
        plt.xticks(rotation=45)
        plt.xlabel('Directions (deg)')
        plt.ylabel('Spike count')
        return
    else:
        return p
````

Plot tuning curve and fit for different neurons. Good candidates to check are 28, 29 or 37. 

````python
neuron = 37

spk_by_dir = spikes[spikes['Neuron']==neuron].groupby(['Dir','Trial'])['stimPeriod'].sum().reset_index()

dirs = spk_by_dir['Dir'].as_matrix()
counts = spk_by_dir['stimPeriod'].as_matrix()

# because we count spikes only when they are present, some zero entries in the count vector are missing
tr_dirs = np.arange(0,nDirs)*deltaDir
for i, dir in enumerate(tr_dirs):
    m = nTrials-np.sum(dirs==dir)
    if m>0:
        dirs = np.concatenate((dirs,np.ones(m)*dir))
        counts = np.concatenate((counts,np.zeros(m)))
    

tuningCurve(counts,dirs, True)
````

## Task 4: Permutation test for direction tuning

Implement a permutation test to quantitatively assess whether a neuron is direction/orientation selective. To do so, project the vector of average spike counts, $m_k=\frac{1}{N}\sum_j x_{jk}$ on a complex exponential with two cycles, $v_k = \exp(\psi i \theta_k)$, where $\theta_k$ is the $k$-th direction of motion in radians and $\psi \in 1,2$ is the fourier component to test (1: direction, 2: orientation). Denote the projection by $q=m^Tv$. The magnitude $|q|$ tells you how much power there is in the $\psi$-th fourier component. 

Estimate the distribution of |q| under the null hypothesis that the neuron fires randomly across directions by running 1000 iterations where you repeat the same calculation as above but on a random permutation of the trials (that is, randomly shuffle the entries in the spike count matrix x). The fraction of iterations for which you obtain a value more extreme than what you observed in the data is your p-value. Implement this procedure in the function ```testTuning()```. 

Illustrate the test procedure for one of the cells from above. Plot the sampling distribution of |q| and indicate the value observed in the real data in your plot. 

How many cells are tuned at p < 0.01?

*Grading: 3 pts*


````python
def testTuning(dirs, counts, psi=1, show=False):
    """
    Plot the data if show is True, otherwise just return the fit.
        count      #trials x 1, contains the spike count during the stimulation period
        dirs       #trials x 1, contains the direction in degrees
        psi        fourier component to test (1 = direction, 2 = orientation)
        show       Boolean, indicates whether a plot is generated
    
        p       p-value
        q       magnitude of second Fourier component
        qdistr  sampling distribution of |q| under the null hypothesis    

    """
    
    nIter = 1000  # Set number of iterations

    idx = np.argsort(dirs)
    ddirs = np.copy(dirs)[idx]
    ccounts = np.copy(counts)[idx]
    udirs = np.unique(ddirs)/180*np.pi   # unique directions in radians

    M = np.sum(ddirs==0)
    N = len(udirs)

    ccounts = np.reshape(ccounts,(N,M))
    

    # Complex exponential projection
    m = np.exp(psi * 1j * udirs) / np.sqrt(N)
    q = np.abs(np.mean(ccounts, axis=1) @ m)

    # Make q-distribution by permuting trials, and get p-value as the percentile of the actual q value
    qdistr = np.zeros(nIter)
    for j in range(nIter):

        # Shuffle trial labels
        ccounts = ccounts.flatten()
        np.random.shuffle(ccounts)
        ccounts = np.reshape(ccounts, (N, M))

        # Project shuffled trials onto complex exponential
        q_shuffle = np.abs(np.mean(ccounts, axis=1) @ m)
        # Insert into q-distribution
        qdistr[j] = q_shuffle

    p = np.mean(qdistr >= q)
    
    if show:
        pp = sns.distplot(qdistr)
        yl = pp.get_ylim()
        plt.plot([q, q],[0, 1],'k')
        plt.xlabel('|q|')
        plt.yticks([])
        plt.ylim(yl)
                
        xl = pp.get_xlim()
        plt.xlim([0, xl[1]])
        plt.text(.6 * xl[1], .9 * yl[1],'p-value: {}'.format(p), fontsize=12)
    else:
        return p, q, qdistr

  
````

Show null distribution for the example cell:

````python
neuron = 28

spk_by_dir = spikes[spikes['Neuron']==neuron].groupby(['Dir','Trial'])['stimPeriod'].sum().reset_index()

dirs = spk_by_dir['Dir'].as_matrix()
counts = spk_by_dir['stimPeriod'].as_matrix()

# because we count spikes only when they are present, some zero entries in the count vector are missing
tr_dirs = np.arange(0,nDirs)*deltaDir
for i, dir in enumerate(tr_dirs):
    m = nTrials-np.sum(dirs==dir)
    if m>0:
        dirs = np.concatenate((dirs,np.ones(m)*dir))
        counts = np.concatenate((counts,np.zeros(m)))
        
testTuning(dirs,counts,show=True,psi=2)
````

Test all cells for orientation and direction tuning

````python
neurons = np.unique(spikes['Neuron'])

spk_by_dir = spikes.groupby(['Neuron','Dir','Trial'])['stimPeriod'].sum().reset_index()


p_dir = np.zeros_like(neurons,dtype='float64')
p_ori = np.zeros_like(neurons,dtype='float64')

for i, n in enumerate(neurons):
    dirs = spk_by_dir[spk_by_dir['Neuron']==n]['Dir'].as_matrix()
    counts = spk_by_dir[spk_by_dir['Neuron']==n]['stimPeriod'].as_matrix()

    # because we count spikes only when they are present, some zero entries in the count vector are missing
    tr_dirs = np.arange(0,nDirs)*deltaDir
    for foo, dir in enumerate(tr_dirs):

        m = nTrials-np.sum(dirs==dir)
        if m>0:
            dirs = np.concatenate((dirs,np.ones(m)*dir))
            counts = np.concatenate((counts,np.zeros(m)))

    p_dir[i] = testTuning(dirs,counts,psi=1)[0]
    p_ori[i] = testTuning(dirs,counts,psi=2)[0]
````
