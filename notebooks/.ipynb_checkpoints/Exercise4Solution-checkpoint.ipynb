## Task 2: Simple deconvolution

It is clear from the above plots that the calcium events happen in relationship to the spikes. As a first simple algorithm implement a deconvolution approach like presented in the lecture in the function `deconv_ca`. Assume an exponential kernel where the decay constant depends on the indicator ($\tau_{OGB}= 0.5 s$, $\tau_{GCaMP}= 0.1 s$). As we know that there can be no negative rates, apply a heavyside function to the output. Plot the kernel as well as an example cell with true and deconvolved spike rates. Scale the signals such as to facilitate comparisons.

*Grading: 3 pts*


```python
def deconv_ca(ca, tau, dt):
    """
    Compute the deconvolved calcium signal 
    sp_hat = ca_sta(ca, sp, tau) computes the deconvolved calcium signal
    using the indicator dependent exponential calicum kernel.
    ---
    parameters:
    : ca:    1 by T
    : tau:   1 by 1
    
    returns: sp_hat: 1 by T
    """

    #dt = 0.01
    t = np.arange(0,3,dt)

    kernel = np.exp(-t/tau)

    sp_hat = signal.deconvolve(ca,kernel)[0]
    sp_hat[sp_hat<0] = 0 

    return sp_hat
```

```python
"""
example plot of kernels
"""
sns.set_context('notebook')

t = np.arange(0,2,0.01)

tau=0.1
kernel = np.exp(-t/tau)
plt.plot(t,kernel, label='tau=0.1')
#plt.axvline(2*tau)

tau=0.5
kernel = np.exp(-t/tau)
plt.plot(t,kernel, label='tau=0.5')
plt.axvline(1*tau, color='orange',ls=':', label='1*tau')
plt.axvline(2*tau, color='orange',ls='--', label='2*tau')
plt.axvline(3*tau, color='orange', label='3*tau')


plt.text(0.6,0.42,'kernel=exp(-t/tau)')
plt.title('exponential_kernel(t,tau)')
plt.xlabel('t [sec]')
plt.legend()
sns.despine()
```

#### Apply the function to the example OGB cell

```python
idx = 4
ca = signal.decimate(ogb_calcium.iloc[:,idx], 4)
sp = signal.decimate(ogb_spikes.iloc[:,idx], 4)
sp[sp<0] = 0 # remove decimation artefacts

tau = .5

sp_hat = deconv_ca(ca, tau, dt=1/25)
```

```python
t = np.arange(0,len(ca)) * 1/25

plt.plot(t,ca)
plt.plot(t[0:len(sp_hat)],3*sp_hat-3)
plt.plot(t,2*sp-5)

plt.xlim((100,200))
plt.xlabel('Time (s)')
plt.yticks([])
plt.legend(['Calcium', 'Inferrred', 'Spikes'])
```

### Apply the function to the example GCamP cell

```python
idx = 5
ca = signal.decimate(gcamp_calcium.iloc[:,idx], 4)
sp = signal.decimate(gcamp_spikes.iloc[:,idx], 4)
sp[sp<0] = 0 # remove decimation artefacts


tau = .1
dt=1/25
sp_hat = deconv_ca(ca, tau,dt)
```

```python
sns.set_context('notebook')

t = np.arange(0,len(ca)) * 1/25

plt.title('Gcamp')
plt.plot(t,ca/1.5)
plt.plot(t[0:len(sp_hat)],sp_hat-4)
plt.plot(t,3*sp-6)

plt.xlim((100,200))
plt.xlabel('Time (s)')
plt.ylim((-6.5,7))
plt.yticks([])
plt.legend(['Calcium', 'Inferrred', 'Spikes'], loc='center left', bbox_to_anchor=(1, 0.5))
```

## Task 4: Evaluation of algorithms

To formally evaluate the algorithms on the two datasets run the deconvolution algorithm and the more complex one on all cells and compute the correlation between true and inferred spike trains. `DataFrames` from the `pandas` package are a useful tool for aggregating data and later plotting it. Create a dataframe with columns

* algorithm
* correlation
* indicator

and enter each cell. Plot the results using `stripplot` and/or `boxplot` in the `seaborn` package.

*Grading: 3 pts*


Evaluate on OGB data

```python
N = ogb_calcium.shape[1]
tau = .5

c1 = np.zeros(N)
c2 = np.zeros(N)
for i in range(0, N):
    ca = signal.decimate(ogb_calcium.iloc[:,i][~np.isnan(ogb_calcium.iloc[:,i])], 4)
    sp = signal.decimate(ogb_spikes.iloc[:,i][~np.isnan(ogb_spikes.iloc[:,i])], 4)
    sp[sp<0] = 0 # remove decimation artefacts
    
    ca = ca[~np.isnan(ca)]
    sp = sp[~np.isnan(sp)]
    
    dt=1/25
    
    sp_hat = deconv_ca(ca, tau,dt)
    c1[i] = np.corrcoef(sp_hat,sp[0:len(sp_hat)])[0,1]

    sp_hat, Cz = oopsi.fast(ca,dt=1/25,iter_max=6)
    c2[i] = np.corrcoef(sp_hat,sp)[0,1]
```

Create OGB dataframe

```python
c = np.concatenate((c1, c2))
algo = np.concatenate((np.tile('deconv',(1,len(c1))),
                       np.tile('oopsi',(1,len(c2))))).flatten()
indicator = np.tile('ogb',(1,len(c))).flatten()
df_ogb = pd.DataFrame({'algorithm': algo, 
        'correlation': c, 
        'indicator': indicator})

df_ogb.head()
```
